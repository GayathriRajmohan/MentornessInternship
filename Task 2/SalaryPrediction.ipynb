**EXPLORATORY DATA ANALYSIS**

Exploratory Data Analysis (EDA) is a statistical approach used to analyze datasets and summarize their main characteristics. It emphasizes using graphical techniques to get a sense of the data before applying formal modeling or hypothesis testing

*   ***Understanding the data*** : EDA helps you summarize the data's key characteristics, like central tendencies and distributions of different variables.
*  ***Finding patterns*** : By visualizing the data, you can discover patterns and relationships between variables that might not be obvious from just looking at raw numbers.

*  ***Identifying issues*** : EDA can help you spot outliers, missing values, or other errors in the data that could affect your analysis.

***Univariate Analysis:***

For Continuous variables - histograms,box plots and mean,median are used

For Categorical variables - Frequency table and bar chart is used

```python
# Histogram, Scatter plot, Mean and Median for Continuous values
# DOJ
# Handle potential non-date values and inconsistencies in format
df['DOJ'] = pd.to_datetime(df['DOJ'], errors='coerce')  # 'coerce' will convert invalid dates to NaT (Not a Time)
df['DOJ_dayofyear'] = df['DOJ'].dt.dayofyear

# Plot histogram (handling NaT values)
plt.figure()
sns.histplot(x='DOJ_dayofyear', data=df.dropna(subset=['DOJ_dayofyear']), kde=True)  # Drop rows with NaT in 'DOJ_dayofyear'
plt.title('Histogram of DOJ Day of Year')
plt.show()

# Plot boxplot (handling NaT values)
plt.figure()
sns.boxplot(x='DOJ_dayofyear', data=df.dropna(subset=['DOJ_dayofyear']))  # Drop rows with NaT in 'DOJ_dayofyear'
plt.title('Boxplot of DOJ Day of Year')
plt.show()

# Calculate mean and median (handling NaT values)
mean_doj = df['DOJ_dayofyear'].mean()
median_doj = df['DOJ_dayofyear'].median()

print("\n\nMean of DOJ Day of Year:", mean_doj)
print("Median of DOJ Day of Year:", median_doj)
```

```python
# Frequency Table and Bar chart for Categorical values
# First name and Last name is not applicable in univariate analysis as it focuses on single variables
# Iterate over the specified columns
for col in ['SEX', 'CURRENT DATE', 'DESIGNATION', 'UNIT']:
    # Calculate frequency table
    freq_table = df[col].value_counts()
    print(f"Frequency Table for {col}:\n{freq_table}\n")

    # Plot bar chart
    plt.figure()
    sns.countplot(x=col, data=df)
    plt.title(f'Bar Chart for {col}')
    plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
    plt.show()
    print("-" * 20)
```

***Bivariate Analysis***

```python
# Bivariate Analysis
# Box plot for SEX vs SALARY
plt.figure()
sns.boxplot(x='SEX', y='SALARY', data=df)
plt.title('Box Plot of Salary by Sex')
plt.show()

# Scatter plot for SEX vs SALARY
plt.figure()
sns.scatterplot(x='SEX', y='SALARY', data=df)
plt.title('Scatter Plot of Salary by Sex')
plt.show()

# Box plot for DESIGNATION vs SALARY
plt.figure()
sns.boxplot(x='DESIGNATION', y='SALARY', data=df)
plt.title('Box Plot of Salary by Designation')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()

# Scatter plot for DESIGNATION vs SALARY
plt.figure()
sns.scatterplot(x='DESIGNATION', y='SALARY', data=df)
plt.title('Scatter Plot of Salary by Designation')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()

# Correlation Coefficient and Scatter Plots

for x_col, y_col in [('AGE', 'SALARY'), ('PAST EXP', 'SALARY'), ('RATINGS', 'SALARY')]:
    # Calculate correlation coefficient
    correlation = df[x_col].corr(df[y_col])
    print(f"\nCorrelation coefficient between {x_col} and {y_col}: {correlation:.2f}\n")

    # Scatter plot
    plt.figure()
    sns.scatterplot(x=x_col, y=y_col, data=df)
    plt.title(f'Scatter Plot of {y_col} vs. {x_col}')
    plt.show()
    print("-" * 20)
```

```python
#Converting date string into year format
import datetime
from datetime import datetime
df['DOJ'] = pd.DatetimeIndex(df['DOJ']).year
df.head()
```

```python
df1 = df.drop(columns=['FIRST NAME', 'LAST NAME', 'CURRENT DATE'])
cat_list=[i for i in df1.select_dtypes("object")]
cat_list
```

```python
#Convert categorical values into binary
for i in cat_list:
    df1[i] = df1[i].factorize()[0]
df1
```

```python
# Calculate correlation matrix
correlation_matrix = df1.corr()

# Plot heatmap of correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
# Splitting Data into Train and Test
X=df1.drop(["SALARY"], axis=1)
Y=df1["SALARY"]
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
!pip install scikit-learn
from sklearn.linear_model import Ridge, Lasso, RidgeCV, LassoCV, ElasticNet, ElasticNetCV, LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score


models = {
    "Ridge": Ridge(),
    "Lasso": Lasso(),
    "RidgeCV": RidgeCV(),
    "LassoCV": LassoCV(),
    "ElasticNet": ElasticNet(),
    "ElasticNetCV": ElasticNetCV(),
    "LinearRegression": LinearRegression(),
    "DecisionTreeRegressor": DecisionTreeRegressor(),
    "KNeighborsRegressor": KNeighborsRegressor(),
    "MLPRegressor": MLPRegressor(),
    "RandomForestRegressor": RandomForestRegressor(),
    "GradientBoostingRegressor": GradientBoostingRegressor(),
    "AdaBoostRegressor": AdaBoostRegressor(),
    "SVR": SVR()
}

results = []

for name, model in models.items():
    model.fit(X_train, Y_train)
    y_pred = model.predict(X_test)

    mse = mean_squared_error(Y_test, y_pred)
    r2 = r2_score(Y_test, y_pred)

    results.append({
        "Model": name,
        "Mean Squared Error": mse,
        "R-squared": r2*100
    })

results_df = pd.DataFrame(results)

# Find the model with the highest R-squared value
best_model_row = results_df.loc[results_df['R-squared'].idxmax()]
best_model_name = best_model_row['Model']
best_model_r2 = best_model_row['R-squared']

print(results_df)
print("\nBest Performing Model:")
print(f"Model: {best_model_name}")
print(f"R-squared: {best_model_r2:.2f}")
```
